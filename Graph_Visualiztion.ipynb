```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
### Matplotlib 사용하기
# 점선 그래프 그리기 
```

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

```python
### 여러 그래프 그릴 준비하기
```

```python
plt.subplot(1,2,1) #(행,렬,순서)
plt.subplot(1,2,2)
plt.show()
```

```python
### Multi Graph 그리기
hist_data = np.random.randint(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data,bins=20)
plt.subplot(2,2,3)
plt.scatter(scat_data,np.arange(30)*np.random.randn(30))
plt.show()
```

```python
### 그래프 선 옵션
```

```python
#그래프를 그릴 때 표시 되는 색이나 마커
#색상 : b(파란색),g(초록색),r(빨간색),y(노란색),k(검은색),w(흰색)
#마커 : o(원),v(역삼각형),^(삼각형),s(네모),+(플러스)..(점),--(점선)
```

```python
plt.plot(data,'b--')
```

```python
plt.plot(data,'k+')
```

```python
### 그래프 사이즈 조절
# plt.figure안에 figsize를 이용하여 가로,세로 길이 조절 가능(inch단위)
```

```python
plt.figure(figsize = (10,5))
plt.plot(data,'k+')
# figure size는 그래프 코드보다 위에 있어야 함
```

```python
### 그래프 겹치기 + legend 달기
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data,'k--',label ='Default')
plt.plot(data,'k-',drawstyle='steps-post',label = 'step-post')
plt.legend()
plt.show()
### subplot을 주지 않았기 때문에 겹쳐서 나옴옴
```

```python
### 이름달기 
```

```python
plt.plot(np.random.randn(30).cumsum())
plt.title('Random Graph')
plt.xlabel('stages')
plt.ylabel('Values')
plt.show()
plt.savefig('saved_graph.png')
```

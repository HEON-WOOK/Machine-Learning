```python
from googleapiclient.discovery import build
from google.colab import drive
import pandas as pd
from IPython.display import display
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/KR_youtube_trending_data.csv',encoding="utf-8")
display(df.head())

youtubeApiKey = 'AIzaSyD0Nbt9ld7mSkAzoGW-zEoMQEZUYi-yizE'
youtube = build('youtube','v3',developerKey=youtubeApiKey)

video_durations = []
cnt = 0

for video_id in df['video_id']:
    try:
        # print(video_id)
        search_response = youtube.videos().list(
            id=video_id,
            part="contentDetails",
        ).execute()

        # print(search_response)
        if len(search_response['items']) == 0:
            video_durations.append((video_id, 0))
        else:
            video_durations.append((video_id, search_response['items'][0]['contentDetails']['duration']))
    except Exception as e:
        print(e)
        print(video_id)
        video_durations.append((video_id,0))
        continue

    if cnt % 1000:
        print(cnt + " is completed..")

# df['duration'] = video_durations

print(video_durations)
# df.to_csv('/content/drive/My Drive/Colab Notebooks/youtube.csv')
```

```python
idList = []
durationList = []
new_df = pd.DataFrame()

for id, dur in video_durations:
    idList.append(id)
    durationList.append(dur)

new_df['id'] = idList
new_df['duration'] = durationList

display(new_df.head())



def convertDuration(durSeries):
    convDur = []
    for dur in durSeries:
        if dur == 0:
            convDur.append(0)
            continue
        
        newDur = ''
        result = 0
        prev = 0
        temp = ''
        for ch in dur:
            if ch == 'H':
                temp +=  '*' + str(3600) + '+'
                result += prev * 3600
                
                newDur = ''
                # newDur += ch + ','
            elif ch == 'M':
                temp += '*' + str(60) + '+'
                result += prev * 60

                newDur = ''
                # newDur += ch + ','
            elif ch == 'S':
                temp += '*' + str(1)
                result += prev * 1

                newDur = ''
                # newDur += ch
            elif ch == 'P' or ch == 'T':
                pass
            else:
                newDur += ch
                temp += ch
                prev = int(newDur)
            # print(newDur)

        # print(">", temp)
        # print(result)
        convDur.append(result)
    return convDur



convDurList = convertDuration(new_df['duration'])

new_df['convDur'] = convDurList
display(new_df)

new_df.to_csv('/content/drive/My Drive/Colab Notebooks/youtube.csv')
```
